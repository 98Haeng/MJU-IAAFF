- 워드투벡 모델 불러오기

```python
from gensim.models import Word2Vec
from gensim.models import KeyedVectors
from tqdm import tqdm_notebook
```

```python
# 이전 그리디 알고리즘.ipnyb파일에서 수행했던 결과
model = Word2Vec.load("final_w2v.model")

# 단어 벡터 확인
word_vectors = model.wv
```

```python
# 단어 리스트 지정 (핵심단어)
word_list = list(filter_df['Word'])
```

```python
# '유사단어1'부터 '유사단어26'까지의 칼럼 이름을 리스트로 만듭니다.
word_columns = [f'유사단어{i}' for i in range(1, 57)]

avg_vector_list = []
for i in tqdm_notebook(range(len(word_list))):
    sim_wordlist = []
    for column in word_columns:
        sim_wordlist.append(df[column][i])

    vector_sum = np.zeros(model.vector_size)
    for word in sim_wordlist:
        word = str(word)
        vector = word_vectors[word]
        vector_sum += vector
    # 원래 단어 벡터 더하기
    origin_vector = word_vectors[word_list[i]]
    vector_sum += origin_vector
    vector_sum = vector_sum / (len(sim_wordlist)+1)
    avg_vector_list.append(vector_sum)
```

```python
len(avg_vector_list)
```

```python
# 각 연도, 벡터들을 하나의 데이터프레임으로 만들기
data_dict = {
    'Word': word_list,
    **{f'vector{i+1}': [vector[i] for vector in avg_vector_list] for i in range(100)}
}

df_avg_vector = pd.DataFrame(data_dict)
```

```python
df_avg_vector['year'] = list(filter_df['Year'])
```
