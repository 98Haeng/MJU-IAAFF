```python
# Library
import pandas as pd
import numpy as np
import os
from tqdm import tqdm

import warnings
warnings.filterwarnings('ignore')

import re
import nltk
```

```python
df = pd.read_csv('연도별단어similar200.csv')
df.head()
```

```python
# Similarity_Scores 칼럼에서 공백을 기준으로 분할
df_split = df['Similarity_Scores'].str.split(' ', expand=True)
```

```python
# 결과를 저장할 새로운 DataFrame
df_final = pd.DataFrame()

# ':'을 기준으로 분할
for i in range(df_split.shape[1]):
    df_temp = df_split[i].str.split(':', expand=True)
    df_final[f'유사단어{i+1}'] = df_temp[0]
    df_final[f'유사도{i+1}'] = df_temp[1]

df_final
```

```python
df_final = pd.concat([df['Word'], df_final], axis=1)
```

```python
df_final
```

```python
df_final.to_csv('단어유사형태200.csv', encoding='utf-8-sig', index=False)
```
