```python
com_df = pd.read_csv('0727community_cout.csv')
com_df.head()
```

```python
com_df['cite_count'] = list(cite_df['cites'])
```

```python
com_df.to_csv('cite반영유사도데이터.csv', encoding='utf-8-sig', index=False)
```

```python
year_total_counts = com_df.groupby('Year')['count'].transform('sum')
# 해당 행의 count 값을 해당 연도 전체 count 합으로 나누어 정규화
com_df['count_norm_paper'] = com_df['count'] / year_total_counts
```

```python
com_df['count_norm_words'] = com_df['count_norm_paper'] / com_df['total_words']
```

```python
com_df.head()
```

```python
year_cite_counts = com_df.groupby('Year')['cite_count'].transform('sum')
# 해당 행의 count 값을 해당 연도 전체 count 합으로 나누어 정규화
com_df['cite_count_norm'] = com_df['cite_count'] / year_cite_counts
```
