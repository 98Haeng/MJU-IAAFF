## 3. Author Keywords Preprocessing

```python
# Author Keywords 확인
new_df = data[data['Author Keywords'].notnull()]
new_df.reset_index(drop=True, inplace=True)
new_df.head()
```

```python
len(new_df)
```

```python
new_df['Author Keywords'][10]
```

```python
# Author Keywords 칼럼에서 괄호 안의 내용은 모두 제거하는 함수
def remove_brackets(text):
    # 괄호와 괄호 안의 내용을 제거하는 정규 표현식
    pattern = r'\([^)]*\)'
    result = re.sub(pattern, '', text)

    # 공백이 두 개 이상 연속으로 있는 경우, 이를 하나의 공백으로 바꿈
    result = re.sub(' +', ' ', result)

    return result
```

```python
# 1차원 형태로 담기 (Author Keywords의 내용)
keywords = []
for i in tqdm(range(len(new_df))):
    for word in new_df['Author Keywords'][i].split('; '):
        keywords.append(word)
```

```python
arr = set(keywords)
print(len(keywords), len(arr))
```

```python
# 모두 소문자로 전환 및 하이픈을 한칸 띄어쓰기로 대체
for i in range(len(keywords)):
    keywords[i] = keywords[i].lower().replace('-',' ')
```

```python
# 1. 키워드 하나를 갖고오고
# 2. 띄어쓰기 기준으로 분할 시킨 다음에
# 3. 띄어쓰기 마지막거만 한 다음에
# 4. 합쳐서 다시 리스트에 집어넣기
# 5. ing가 들어간 경우는 그대로 놔두기
lemma = WordNetLemmatizer()
keywords_changed = []

for i in tqdm(range(len(keywords))):
    w = keywords[i]
    for i in range(len(w.split(' '))):
        a = w.split(' ')
        if i == len(a)-1:
            word = a[i] # 마지막 단어만 추출
            word = lemma.lemmatize(word,'n')# 단수형태로 변환
#             print(word)
            a[-1] = word
            a = " ".join(a)# 다시 합치기
    keywords_changed.append(a)
```

#### 변화 결과 비교
- 소문자 변환
- 복수형을 단수로 변환

```python
# 단어 빈도 수 확인
# 여기서 빈도수는 Author Keywords 기준!
c = Counter(keywords_changed)
most_keywords_list = list(c)
dic_c = dict(c)
keys = list(dic_c.keys())
vals = []
for i in range(len(keys)):
    vals.append(dic_c[keys[i]])
    
words_list = pd.DataFrame([keys,vals]).T
words_list.columns = ['word','count']

words_list.sort_values(by=['count','word'], ascending=False, inplace=True)
```

```python
words_list.head()
```

```python
# 괄호가 들어간 단어 제거
words_list['word'] = words_list['word'].progress_apply(remove_brackets)
```

```python
# 단어 리스트 저장
words_list.to_csv('Data/Preprocess_Data/Author_Keywords.csv', encoding='utf-8-sig', index=False)
```

## 4. Abstract Preprocessing

```python
# 대소문자 변환 및 하이픈 공백으로 대체
data['Abstract'] = data['Abstract'].str.lower().str.replace('-', ' ')
```

```python
# 마침표와 콤마 제거
import re

def remove_punctuation(text):
    # 정규표현식 패턴을 사용하여 '.'와 ','를 제거
    cleaned_text = re.sub('[.,]', '', text)
    return cleaned_text

# 데이터프레임의 모든 열에 대해 함수를 적용합니다.
data['Abstract'] = data['Abstract'].map(remove_punctuation)
```

```python
# 괄호 안에 있는 모든 내용 제거
data['Abstract'] = data['Abstract'].progress_apply(remove_brackets)
```

```python
# 인덱스 초기화 한번 해주기
data.reset_index(drop=True, inplace=True)
```

```python
# Abstracts 복수를 단수로 변환
abstracts_changed = []
for i in tqdm(range(len(data))): # ab의 길이 = in_abs의 길이
    sentence = data['Abstract'][i]
    a = sentence.split(' ')
    for k in range(len(a)):
        a[k] = lemma.lemmatize(a[k],'n')
    a2 = " ".join(a)
    abstracts_changed.append(a2)
```

```python
data['Abstract2'] = abstracts_changed
```
