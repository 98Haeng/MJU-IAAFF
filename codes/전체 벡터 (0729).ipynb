```python
from gensim.models import Word2Vec
from gensim.models import KeyedVectors
from tqdm import tqdm_notebook

# 이전 그리디 알고리즘.ipnyb파일에서 수행했던 결과
model = Word2Vec.load("final_w2v.model")

# 단어 벡터 확인
word_vectors = model.wv
```

```python
filter_df = pd.read_csv('단어유사형태200전처리.csv')
```

```python
a = filter_df[filter_df['Year']==2023]
a.reset_index(drop=True, inplace=True)
```

```python
# 단어 리스트 지정 (핵심단어)
word_list = list(a['Word'])
```

```python
word_list = []
word_list.append(a['Word'][0])
for i in range(1,57):
    word_list.append(a['유사단어{}'.format(i)][0])
```

```python
word_list = word_list[:35]
```

```python
word_list
```

```python
from tqdm import tqdm
```

```python
word_vector_list = []
for i in tqdm(range(len(word_list))):
    word = str(word_list[i])
    vector = word_vectors[word]
    word_vector_list.append(vector)
```

```python
word_vector_list
```

```python
a = pd.DataFrame(word_vector_list)
a['word']= word_list
a.to_csv('샘플벡터.csv', index=False, encoding='utf-8-sig')
```

```python
df = pd.read_csv('Word_AverageVector2.csv')
df.head()
```

```python
df = df[df['year'] >= 2022]
```

```python
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
vector_columns = [f'vector{i}' for i in range(1, 101)]
df['vector'] = df.apply(lambda row: [row[col] for col in vector_columns], axis=1)
```

```python
df
```

```python
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
```

```python
np.array(list(df['vector']))
```

```python

# t-SNE를 사용하여 100차원 데이터를 2차원으로 축소
tsne = TSNE(n_components=2, random_state=42)
data_2d = tsne.fit_transform(np.array(list(df['vector'])))

# 시각화
plt.figure(figsize=(8, 6))
for year in df['year'].unique():
    mask = df['year'] == year
    plt.scatter(data_2d[mask, 0], data_2d[mask, 1], label=str(year), alpha=0.7)
plt.xlabel('t-SNE Dimension 1')
plt.ylabel('t-SNE Dimension 2')
plt.title('t-SNE Visualization of 2-dimensional Data with Colors Based on Year')
plt.legend()
plt.show()
```

```python
from sklearn.decomposition import PCA
# 시각화를 위해 PCA를 사용하여 2차원으로 축소
pca = PCA(n_components=2, random_state=42)
data_2d = pca.fit_transform(np.array(list(df['vector'])))

# 시각화
plt.figure(figsize=(8, 6))
for year in df['year'].unique():
    mask = df['year'] == year
    plt.scatter(data_2d[mask, 0], data_2d[mask, 1], label=str(year), alpha=0.7)
plt.xlabel('PCA Dimension 1')
plt.ylabel('PCA Dimension 2')
plt.title('PCA Visualization of 100-dimensional Data with Colors Based on Year')
plt.legend()
plt.show()
```
